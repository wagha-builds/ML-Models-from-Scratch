# Importing the libraries.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Adaline Class (FBGD)

```python
class AdalineSGD:

  def __init__(self, eta=0.01, n_iter=10, shuffle=True, random_state=1):
    """Initializing hyperparameters"""

    self.eta = eta
    self.n_iter = n_iter
    self.shuffle = shuffle
    self.random_state = random_state
    self.w_initialized = False

  def fit(self, X, y):
    """Fit training data"""

    self.initialize_weights_(X.shape[1])
    self.losses_ = []

    for i in range(self.n_iter):
      if self.shuffle:
        X, y = self.shuffle_(X, y)
      losses = []
      for xi, target in zip(X, y):
        losses.append(self.update_weights_(xi, target))
      avg_loss = np.mean(losses)
      self.losses_.append(avg_loss)

    return self

  def partial_fit(self, X, y):
    """Fit training data without reinitializing the weights"""

    if not self.w_initialized:
      self.w_initialized(X.shape[1])
      if y.ravel().shape[0] > 1 :
        for xi, target in zip(X, y):
          self.update_weights_(xi, target)
      else:
        self.update_weights_(X, y)

      return self

  def initialize_weights_(self, m):
    """Initializing parameters"""

    self.rgen = np.random.RandomState(self.random_state)
    self.w_ = self.rgen.normal(loc=0.0, scale=0.01, size=m)
    self.b_ = np.float64(0.)
    self.w_initialized = True

  def shuffle_(self, X, y):
    """Shuffle training data"""

    r = self.rgen.permutation(len(y))
    return X[r], y[r]

  def update_weights_(self, xi, target):
    """Updating parameters"""

    output = self.activation(self.net_input(xi))
    error = (target - output)
    self.w_ += self.eta * 2.0 * xi * (error)
    self.b_ += self.eta * 2.0 * (error)
    loss = error**2
    return loss

  def net_input(self, X):
    return np.dot(X, self.w_) + self.b_

  def activation(self, X):
    return X

  def predict(self, X):
    return np.where(self.activation(self.net_input(X)) >= 0.5, 1, 0)
```

# Target vector (100 examples with only 2 classes)

```python
y = df.iloc[0:100, 4].values # First 100 examples.
y = np.where(y == 'Iris-setosa', 0, 1) # Converting to binary class labels.
```

# Feature matrix (Only 2 features)

```python
X = df.iloc[0:100, [0, 2]].values # 1st and 3rd feature column.
```

# Standardized Feature Scaling

```python
X_std = np.copy(X)
X_std[:, 0] = (X[:, 0] - X[:, 0].mean()) / X[:, 0].std()
X_std[:, 1] = (X[:, 1] - X[:, 1].mean()) / X[:, 1].std()
```

## Adaline with Standardized Features

```python
ada_sgd = AdalineSGD(n_iter=15, eta=0.01)
ada_sgd.fit(X_std, y)
```

# Scatterpoint training examples & Decision Region Plotting Function

```python
from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, resolution=0.02):

  # Defines markers, colors and a colormap.
  markers = ('o', 's', '^', 'v', '<')
  colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
  cmap = ListedColormap(colors[:len(np.unique(y))])

  # Min. and max. values of two features.
  x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
  x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1

  # Grid of feature values.
  xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                         np.arange(x2_min, x2_max, resolution))

  # Predict from 'ppn' model and Reshape into grid of same dimensions xx1 and xx2.
  lab = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
  lab = lab.reshape(xx1.shape)

  # contourf() to color decision regions
  plt.contourf(xx1, xx2, lab, alpha=0.3, cmap=cmap)
  plt.xlim(xx1.min(), xx1.max())
  plt.ylim(xx2.min(), xx2.max())

  for idx, cl in enumerate(np.unique(y)):
    plt.scatter(x=X[y == cl, 0],
                y=X[y == cl, 1],
                alpha=0.8,
                c=colors[idx],
                marker=markers[idx],
                label=f'Class {cl}',
                edgecolor='black')
```

# Visualizing Adaline's Decision Regions

```python
plot_decision_regions(X_std, y, classifier=ada_sgd)
plt.title('Adaline - Stochastic Gradient Descent')
plt.xlabel('Sepal Length [standardized]')
plt.ylabel('Petal Length [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

# MSE v/s No. of Epochs (Error Plot)

```python
plt.plot(range(1, len(ada_sgd.losses_) + 1), ada_sgd.losses_, marker='o')
plt.xlabel('Epochs')
plt.ylabel('MSE')
plt.tight_layout()
plt.show()
```
